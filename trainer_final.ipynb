```python
import numpy as np
import os
import json
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
with open("var_store.json", 'r') as file:
    var_store = json.load(file)
```

```python
path = os.path.abspath("")
```

```python
path
```

```python
import shutil
shutil.rmtree("Keypoints")
```

```python
features  = np.array(var_store['features'])
avg_vid_count = 38
avg_fr_len = 35
```

```python
label_map = {label:num for num,label in enumerate(features)}
```

```python
label_map
```

```python
sequences, labels =  [], []
for index, feature in enumerate(features):
    for sequence in range(avg_vid_count):
        window = []
        frame_path = os.path.join(path,"Keypoints", str(index), str(sequence))
        for frame_num in range(avg_fr_len):
            res = np.load(os.path.join(frame_path,'{}.npy'.format(frame_num)))
            window.append(res)
        sequences.append(window)
        labels.append(label_map[feature])
```

```python
np.array(sequences).shape
```

```python
np.array(labels).shape
```

```python
y = to_categorical(labels).astype(int)
X = np.array(sequences)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.05)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv1D,LSTM, MaxPooling1D
# from tensorflow.keras.callbacks import TensorBoard
```

```python

model = Sequential()

# CNN layers
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(35, 1086)))
model.add(MaxPooling1D(pool_size=2))

# LSTM layers
model.add(LSTM(64, return_sequences=True, activation='relu'))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))

# Dense layers
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(features.shape[0], activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train, epochs= 1500)
```
